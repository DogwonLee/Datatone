# 1. 데이터셋 컬럼명 한글화

## 1-1. events 컬럼명 한글화

```python
# events 데이터 불러오기
import pandas as pd
events = pd.read_csv("/content/events1.csv")
events
```

```python
# 컬럼명 변경하기
events.columns = [
    "사용자ID",
    "세션_식별자",
    "국가",
    "기기",
    "타입",
    "상품고유번호",
    "이벤트발생시간"
]

# 확인
events.head()
```

```python
# 파일 다시 저장하기
events.to_csv("/content/events1.csv", index=False, encoding="utf-8-sig")
```

```python
events
```

## 1-2. items 컬럼명 한글화

```python
# items 데이터 불러오기
items = pd.read_csv("/content/items.csv")
items
```

```python
# 컬럼명 변경
items.columns = [
    "상품ID",
    "상품이름",
    "상품브랜드",
    "옵션",
    "상품카테고리",
    "상품가격"
]

# 확인
items.head()
```

```python
# 파일 다시 저장하기
items.to_csv("/content/items.csv", index=False, encoding="utf-8-sig")
```

```python
items
```

## 1-3. users 컬럼명 한글화

```python
# users 데이터 불러오기
users = pd.read_csv("/content/users.csv")
users
```

```python
# 컬럼명 변경
users.columns = [
    "사용자ID",
    "누적구매금액",
    "시간"
]

# 확인
users.head()
```

```python
# 파일 다시 저장하기
users.to_csv("/content/users.csv", index=False, encoding="utf-8-sig")
```

```python
users
```

# 2. 데이터셋 병합하기

## 2-1. events 데이터, users 데이터 병합

```python
# '사용자ID'기준 조인
events_users = pd.merge(events, users, how="left", left_on="사용자ID", right_on="사용자ID")
```

```python
events_users
```

## 2-2. events_users 데이터, items 데이터 병합

```python
# '상품에 대한 고유 정보' 기준 조인
total = pd.merge(events_users, items, how="left", left_on="상품고유번호", right_on="상품ID")
total
```

## 2-3. 최종 데이터셋 저장

```python
# 파일 저장하기
total.to_csv("/content/통합데이터셋(dataton).csv", index=False, encoding="utf-8-sig")
```
